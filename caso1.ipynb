```
import matplotlib as plt
from scipy.cluster.hierarchy import dendrogram, linkage
import numpy as np
import matplotlib.pyplot as plt


np.random.seed(123)
a=np.random.multivariate_normal([10,0], [[2,1],[1,4]], size=[100,])
b=np.random.multivariate_normal([0,30], [[2,1],[1,4]], size=[50,])
a[:6]
b[:6]

X=np.concatenate((a,b))
print(X.shape)


```

```
plt.scatter(X[:,0],X[:,1])
plt.show()
```

```
distancias=linkage(X)
distancias[:6]
```

```
plt.figure(figsize=(30,10))
dendrogram(distancias)
plt.show
```

```
plt.figure(figsize=(30,10))
dendrogram(distancias,truncate_mode="lastp",p=10 )
plt.show()
```
